```python
# Data Wraggling
import numpy as np
import pandas as pd
from numpy import *
# Statistics
from math import sqrt
# Visualisation
import seaborn as sns
import matplotlib.pyplot as plt
# Scikit learn
from sklearn import datasets
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report, confusion_matrix
from sklearn.metrics import precision_recall_fscore_support as score
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
# Keras 
import tensorflow
from tensorflow import keras
from keras.layers import Dense
from keras.models import Sequential
from tensorflow.keras.optimizers import SGD, Adam
from tensorflow.keras.callbacks import Callback
from keras.callbacks import LearningRateScheduler

def label_col (row, col_name):
    if row[col_name]>=0 and row[col_name]<=7:
        return "C1"
    elif row[col_name]>=8 and row[col_name]<=10:
        return "C2"
    elif row[col_name]>=11 and row[col_name]<=15:
        return "C3"
    return "C4"

def normalize(df):
    # define min max scaler
    scaler = MinMaxScaler()
    # transform data
    scaled = scaler.fit_transform(df.values)
    # convert the array back to a dataframe
    result = pd.DataFrame(scaled, columns=["sex", "length", "diameter", "height", "whole weight", "shucked weight", "viscera weight", "shell weight"])
    return result

def load_data(data_url, normalise):
    column_names = ["sex", "length", "diameter", "height", "whole weight", "shucked weight", "viscera weight", "shell weight", "rings"]

    data = pd.read_csv(data_url, sep=',', names=column_names)
    #data.head()
    
    df_org = data.copy()
    
    # Convert categorical feature to numerical one
    encoder = LabelEncoder()
    encoder.fit(data['sex'])
    data['sex'] = encoder.transform(data['sex'])

    # Convert rings to categories
    #data['age'] = data.apply(lambda _: '', axis=1)
    data['age'] = data.apply(lambda row: label_col(row, 'rings'), axis=1)
    n_class = len(unique(data.age.values))
    data_X = data.drop(['rings', 'age'], axis=1)
    data_y = data[['age']]
    dataset = pd.concat([data_X, data_y], axis=1)
    #dataset.head()

    # One hot encoding the sex attribute
    df_dummies = pd.get_dummies(dataset['sex'], prefix = "sex_")

    # Inserting dummy columns
    for column in df_dummies.columns:
        dataset[column] = df_dummies[column]

    # Dropping the original column
    dataset = dataset.drop(columns = ['sex'])

    # One hot encoding the age attribute
    df_dummies = pd.get_dummies(dataset['age'])

    # Inserting dummy columns
    for column in df_dummies.columns:
        dataset[column] = df_dummies[column]

    # Dropping the original column
    dataset = dataset.drop(columns = ['age'])

    # Defining the input and output columns to separate the dataset in the later cells.
    input_columns = dataset.columns.tolist()
    input_columns.remove('C1')
    input_columns.remove('C2')
    input_columns.remove('C3')
    input_columns.remove('C4')
    output_columns = ['C1', 'C2', 'C3', 'C4']

    # Splitting into train and test set by 50-50 split ratio
    train_df, test_df = train_test_split(dataset, test_size = 0.4)

    # Splitting into X (input) and y (output)
    X_train, y_train = np.array(train_df[input_columns]), np.array(train_df[output_columns])
    X_test, y_test = np.array(test_df[input_columns]), np.array(test_df[output_columns])

    if normalise == True:
        ss = StandardScaler()
        X_train = ss.fit_transform(X_train)
        X_test = ss.transform(X_test)        
    else:
        pass

    # Save scaled data as .csv file
    abalone_train = np.concatenate((X_train,y_train),axis=1)
    np.savetxt("data/classification/abalone_train.csv", abalone_train, delimiter=",")
    abalone_test = np.concatenate((X_test,y_test),axis=1)
    np.savetxt("data/classification/abalone_test.csv", abalone_test, delimiter=",")
    abalone = np.concatenate((abalone_train, abalone_test))
    np.savetxt("data/classification/abalone.csv", abalone, delimiter=",")
    
    abalone_df = pd.DataFrame(data=abalone, columns=["length","diameter","height","whole weight",
                                                 "shucked weight","viscera weight","shell weight",
                                                 "sex__F","sex__I","sex__M","C1","C2","C3","C4"])
    #abalone_df.head()

    return dataset, df_org, n_class, X_train, X_test, y_train, y_test

def countplot(df, problem_type):
    if problem_type == 0:   # classification problem
        sns.countplot(x='age', data=df)  # using dataset
    if problem_type == 1:   # regression problem
        sns.countplot(x='sex', data=df)  # using df_org
    plt.savefig('figure/cntplt.png')
    plt.show()
    plt.clf()

def corr_plot(df):
    # Develop a correlation map using a heatmap
    corr_matrix = df.corr().round(2)
    plt.figure(figsize=(10,10))
    ax = sns.heatmap(corr_matrix, vmin=-1, center=0, annot=True, cmap='BrBG')
    plt.savefig('figure/corr_mat.png')
    plt.show()
    plt.clf()

def boxplt(df):
    sns.set_theme(style="darkgrid")
    sns.boxplot(x=df["sex"], y=df['rings'])
    plt.savefig('figure/boxplt.png')
    plt.show()
    plt.clf()

def boxenplot(df):
    # Categorical features
    temp = pd.concat([df['rings'], df['sex']], axis=1)
    f, ax = plt.subplots(figsize=(8, 6))
    fig = sns.boxenplot(x='sex', y="rings", data=df)
    fig.axis(ymin=0, ymax=30)
    plt.savefig('figure/boxen.png')
    plt.show()
    plt.clf()

def pairgrid(df):
    g = sns.PairGrid(df, diag_sharey=False, corner=True)
    g.map_lower(sns.scatterplot)
    g.map_diag(sns.kdeplot)
    plt.savefig('figure/pairgrid.png')
    plt.show()
    plt.clf()

def hist_plot(df):
    # Create histograms of all features
    fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(16,10))
    df.plot(kind='hist', subplots=True, ax=axes, alpha=0.5, bins=30)
    plt.savefig('figure/hist.png')
    plt.show()
    plt.clf()    
    
def keras_nn(X_train, X_test, y_train, y_test, hidden, outputs, model_type, learn_rate):
    '''Neural network.
    :param hidden: Number of neurons in the hidden layer.
    :param model_type: Optimizer options: 
        0 -single hidden layer using SGD, 1 -single hidden layer using Adam, 2 -two hidden layers using SGD
        3 -single hidden layer using Adam, 4 -two hidden layers using Adam.
    :param learn_rate: Tuning parameter in an optimization algorithm.
    :returns:
        mse_train, mse_test, rmse, r2, residuals, history.
    '''    
    if model_type == 0:   # Task 2: effect of number of hidden neurons -single hidden layer using Adam
        # Create the Sequential model
        model = Sequential([
            keras.layers.Dense(hidden, input_dim=X_train.shape[1], activation=keras.layers.LeakyReLU(alpha=0.01), kernel_initializer='random_normal'),
            keras.layers.Dense(outputs, activation='softmax')
        ])
        # Specify the loss function and the optimizer after a model is created
        model.compile(loss="categorical_crossentropy", optimizer='adam', metrics=['accuracy'])
            
    elif model_type == 1:   # Task 3: effect of learning rate -single hidden layer using SGD
        # Create the Sequential model
        model = Sequential([
            keras.layers.Dense(hidden, input_dim=X_train.shape[1], activation=keras.layers.LeakyReLU(alpha=0.01), kernel_initializer='random_normal'),
            keras.layers.Dense(outputs, activation='softmax')
        ])
        # Specify the loss function and the optimizer after a model is created
        sgd = SGD(learning_rate=learn_rate)
        model.compile(loss="categorical_crossentropy", optimizer=sgd, metrics=['accuracy'])
        
    elif model_type == 2:   # Task 4: two hidden layers using SGD
        # Create the Sequential model
        model = Sequential([
            keras.layers.Dense(hidden, input_dim=X_train.shape[1], activation=keras.layers.LeakyReLU(alpha=0.01), kernel_initializer='random_normal'),
            keras.layers.Dense(hidden, activation="relu", kernel_initializer='random_normal'),
            keras.layers.Dense(outputs, activation='softmax')
        ])
        # Specify the loss function and the optimizer after a model is created
        sgd = SGD(learning_rate=learn_rate)
        model.compile(loss="categorical_crossentropy", optimizer=sgd, metrics=['accuracy'])
        
    elif model_type == 3:   # Task 5: Adam vs. SGD -single hidden layer using Adam
        # Create the Sequential model
        model = Sequential([
            keras.layers.Dense(hidden, input_dim=X_train.shape[1], activation=keras.layers.LeakyReLU(alpha=0.01), kernel_initializer='random_normal'),
            keras.layers.Dense(outputs, activation='softmax')
        ])
        # Specify the loss function and the optimizer after a model is created 
        adam = Adam(learning_rate=learn_rate)
        model.compile(loss="categorical_crossentropy", optimizer=adam, metrics=['accuracy'])
        
    elif model_type == 4:   # Task 5: Adam vs. SGD -two hidden layers using Adam
        # Create the Sequential model
        model = Sequential([
            keras.layers.Dense(hidden, input_dim=X_train.shape[1], activation=keras.layers.LeakyReLU(alpha=0.01), kernel_initializer='random_normal'),
            keras.layers.Dense(hidden, activation="relu", kernel_initializer='random_normal'),
            keras.layers.Dense(outputs, activation='softmax')
        ])
        # Specify the loss function and the optimizer after a model is created 
        adam = Adam(learning_rate=learn_rate)
        model.compile(loss="categorical_crossentropy", optimizer=adam, metrics=['accuracy'])
            
    else:
        print('no model')
        
    # Fit model
    history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=3000, batch_size=10, verbose=0)
    #history = model.fit(X_train, y_train, validation_split=0.3, epochs=3000, batch_size=10, verbose=0)

    # Evaluate the model
    # https://keras.io/api/models/model_training_apis/
    loss_train, acc_train = model.evaluate(X_train, y_train, verbose=0)
    loss_test, acc_test = model.evaluate(X_test, y_test, verbose=0)
    #print('Train: %.3f, Test: %.3f' % (train_acc, test_acc))
    
    # Predit
    y_proba = model.predict(X_test)
    #y_pred = np.argmax(model.predict(X_test), axis=1)
    y_pred = np.argmax(y_proba,axis=1)
    y_test_arg=np.argmax(y_test,axis=1)
    acc = accuracy_score(y_test_arg, y_pred)
    interval = 1.96 * sqrt((acc * (1 - acc)) / 50)    
    
    return history, acc, interval, acc_train, acc_test, y_pred


def result_present(y_test, y_pred, history, acc, interval, acc_train, acc_test, task, run_num, fname):
    # Plot metrics
    pd.DataFrame(history.history).plot(figsize=(8, 5))
    plt.grid(True)
    plt.title('Loss&Accuracy_{}_{}_run_{}.png'.format(task, run_num, fname))
    plt.xlabel('Epoch')
    plt.ylabel('Loss&Accuracy')
    plt.gca().set_ylim(0, 1) # set the vertical range to [0-1]
    plt.savefig('figure/classification/{}/Loss&Accuracy_{}_run_{}.png'.format(task, run_num, fname))
    plt.show()
    #plt.clf()

    print('accuracy_train: %.3f' % acc_train)
    print('accuracy_test: %.3f' % acc_test)
    print('Accuracy: %.3f' % acc)
    print('CI of accuracy: %.3f' % interval)


# Driver code
def main():

    # Task 1: Analyse and visualise the given data sets by reporting the distribution of class, distribution of features and any relevant visualisation
    
    normalise = True   # normalise the data
    
    # Get all dataset and subsets
    data_url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data'
    dataset, df_org, n_class, X_train, X_test, y_train, y_test = load_data(data_url, normalise)
    
    # Countplot w.r.t. the categorical feature 'sex'
    countplot(dataset, problem_type)
    # Boxenplot to display the distribution of data based on min, Q1, median, Q3, and max values
    boxenplot(df_org)
    # Correlation heatmap to display the dependence between features and response variables
    corr_plot(dataset)
    # Histograms to display the distribution of continuous numerical features
    hist_plot(dataset)
    # Subplot grid for plotting pairwise relationships in a dataset
    pairgrid(dataset)

    max_expruns = 10    # run 10 experiments for task 2-5
    outputs = n_class   # multi-classification
    
    # Task 2: Investigate the effect of the number of hidden neurons (e.g. 5, 10, 15, 20) for a single hidden layer
    task = 'task2'    
    
    hidden_lst = [5, 10, 15, 20]
    num_hidden = len(hidden_lst)
    col_names = ['5-neuron', '10-neuron', '15-neuron', '20-neuron']
    learn_rate = 0.01
    model_type = 0   # using optimizer Adam to compare effect of number of hidden neurons
    acc_mx = np.zeros((max_expruns, num_hidden))
    ci_mx = np.zeros((max_expruns, num_hidden))
    
    for run_num in range(max_expruns):
        print("\nTask 2: Experiment {} running...\n".format(run_num))
        
        for hidden in hidden_lst:
            hidden_idx = hidden_lst.index(hidden)
            history, acc, interval, acc_train, acc_test, y_pred = keras_nn(X_train, X_test, y_train, y_test, hidden, outputs, model_type, learn_rate)            

            acc_mx[run_num, hidden_idx] = acc
            ci_mx[run_num, hidden_idx] = interval
            
            fname = str(hidden) + '-neuron' 
            result_present(y_test, y_pred, history, acc, interval, acc_train, acc_test, task, run_num, fname)
            
    acc_df = pd.DataFrame(acc_mx, columns=col_names)
    ci_df = pd.DataFrame(ci_mx, columns=col_names)
    acc_df.to_csv("data/classification/task2_acc.csv", encoding='utf-8', index=False)
    ci_df.to_csv("data/classification/task2_CI.csv", encoding='utf-8', index=False)
    #print("\n Task2 Accuracy score: \n", acc_df)
    #print("\n Task2 CI: \n", ci_df)
    
    acc_mean = acc_mx.mean(axis=0)
    acc_std = acc_mx.std(axis=0)
    
    ci_mean = ci_mx.mean(axis=0)
    ci_std = ci_mx.std(axis=0)    
    
    np.savetxt('data/classification/task2_results.txt', (acc_mean, acc_std, ci_mean, ci_std), fmt='%1.5f')

    print("\n Task2 Mean Acc: ", acc_mean)
    print("\n Task2 std of Acc: ", acc_std)
    print("\n Task2 Mean CI: ", ci_mean)
    print("\n Task2 std of CI: ", ci_std)
    
       
    # Task 3: Investigate the effect of learning rate (in case of SGD) for the selected data set (using the optimal number of hidden neurons).
    task = 'task3'    
    
    lr_lst = list(np.arange(0.01, 0.1, 0.02))
    num_lr = len(lr_lst)
    col_names = ['lr=0.01', 'lr=0.03', 'lr=0.05', 'lr=0.07', 'lr=0.09']
    model_type = 1   # using optimizer SGD to compare effect of learning rate
    hidden = 5      # based on result of Task 2, 5 hidden neurons outperformed
    acc_mx = np.zeros((max_expruns, num_lr))
    ci_mx = np.zeros((max_expruns, num_lr))
    
    for run_num in range(max_expruns):
        print("\nTask 3: Experiment {} running...\n".format(run_num))
        
        for learn_rate in lr_lst:
            lr_idx = lr_lst.index(learn_rate)
            history, acc, interval, acc_train, acc_test, y_pred = keras_nn(X_train, X_test, y_train, y_test, hidden, outputs, model_type, learn_rate)

            acc_mx[run_num, lr_idx] = acc
            ci_mx[run_num, lr_idx] = interval
            
            fname = 'learn_rate_' + str(learn_rate.round(2))
            result_present(y_test, y_pred, history, acc, interval, acc_train, acc_test, task, run_num, fname)
    
    acc_df = pd.DataFrame(acc_mx, columns=col_names)
    ci_df = pd.DataFrame(ci_mx, columns=col_names)
    acc_df.to_csv("data/classification/task3_acc.csv", encoding='utf-8', index=False)
    ci_df.to_csv("data/classification/task3_CI.csv", encoding='utf-8', index=False)
    
    acc_mean = acc_mx.mean(axis=0)
    acc_std = acc_mx.std(axis=0)
    
    ci_mean = ci_mx.mean(axis=0)
    ci_std = ci_mx.std(axis=0)    
    
    np.savetxt('data/classification/task3_results.txt', (acc_mean, acc_std, ci_mean, ci_std), fmt='%1.5f')

    print("\n Task3 Mean Acc: ", acc_mean)
    print("\n Task3 std of Acc: ", acc_std)
    print("\n Task3 Mean CI: ", ci_mean)
    print("\n Task3 std of CI: ", ci_std)
    

    # Task 4: Investigate the effect on a different number of hidden layers (1, 2) with the optimal number of hidden neurons (from Part 4).
    task = 'task4'
    
    model_type = 2      # two hidden layers with SGD
    learn_rate = 0.03   # based on result of Task 3, learning rate 0.03 outperformed
    hidden = 5         # based on result of Task 2, 5 hidden neurons outperformed    
    acc_lst = np.zeros(max_expruns)
    ci_lst = np.zeros(max_expruns)
    
    for run_num in range(max_expruns):
        print("\nTask 4: Experiment {} running...\n".format(run_num))
        acc_lst = np.zeros(max_expruns) 
        history, acc, interval, acc_train, acc_test, y_pred = keras_nn(X_train, X_test, y_train, y_test, hidden, outputs, model_type, learn_rate)
        
        fname = '2-hidden-layer'
        result_present(y_test, y_pred, history, acc, interval, acc_train, acc_test, task, run_num, fname)
    
        acc_lst[run_num] = acc
        ci_lst[run_num] = interval
    
    np.savetxt('data/classification/task4_acc.csv', acc_lst, delimiter=',')
    np.savetxt('data/classification/task4_CI.csv', ci_lst, delimiter=',')
    
    print("Task4 Accuracy score: \n", acc_lst)
    print("\nTask4 CI: \n", ci_lst)
    
    acc_mean = np.mean(acc_lst)
    acc_std = np.std(acc_lst)

    ci_mean = np.mean(ci_lst)
    ci_std = np.std(ci_lst)
    
    np.savetxt('data/classification/task4_results.txt', (acc_mean, acc_std, ci_mean, ci_std), fmt='%1.5f')

    print("\n Task4 Mean Acc: ", acc_mean)
    print("\n Task4 std of Acc: ", acc_std)
    print("\n Task4 Mean CI: ", ci_mean)
    print("\n Task4 std of CI: ", ci_std)
       
    
    # Task 5_1: Investigate the effect of Adam and SGD on training and test performance using single hidden layer.
    task = 'task5_1'
    
    model_type = 3      # single hidden layer using Adam to compare to SGD done in task 3 
    learn_rate = 0.03   # based on result of Task 3, learning rate 0.03 outperformed    
    hidden = 5         # based on result of Task 2, 5 hidden neurons outperformed    
    acc_lst = np.zeros(max_expruns)
    ci_lst = np.zeros(max_expruns)
    
    for run_num in range(max_expruns):
        print("\nTask 5_1: Experiment {} running...\n".format(run_num))
        history, acc, interval, acc_train, acc_test, y_pred = keras_nn(X_train, X_test, y_train, y_test, hidden, outputs, model_type, learn_rate)
        
        fname = 'adam_vs_sgd_1'
        result_present(y_test, y_pred, history, acc, interval, acc_train, acc_test, task, run_num, fname)
    
        acc_lst[run_num] = acc
        ci_lst[run_num] = interval
        
    np.savetxt('data/classification/task5_1_acc.csv', acc_lst, delimiter=',')
    np.savetxt('data/classification/task5_1_CI.csv', ci_lst, delimiter=',')
    
    print("Task5_1 Accuracy score: \n", acc_lst)
    print("\nTask5_1 CI: \n", ci_lst)
    
    acc_mean = np.mean(acc_lst)
    acc_std = np.std(acc_lst)

    ci_mean = np.mean(ci_lst)
    ci_std = np.std(ci_lst)
    
    np.savetxt('data/classification/task5_1_results.txt', (acc_mean, acc_std, ci_mean, ci_std), fmt='%1.5f')

    print("\n Task5_1 Mean Acc: ", acc_mean)
    print("\n Task5_1 std of Acc: ", acc_std)
    print("\n Task5_1 Mean CI: ", ci_mean)
    print("\n Task5_1 std of CI: ", ci_std)
    
    
    # Task 5_2: Investigate the effect of Adam and SGD on training and test performance using two hidden layers.
    task = 'task5_2'
    
    model_type = 4      # single hidden layer using Adam to compare to SGD done in task 3 
    learn_rate = 0.03   # based on result of Task 3, learning rate 0.03 outperformed    
    hidden = 5         # based on result of Task 2, 5 hidden neurons outperformed
    acc_lst = np.zeros(max_expruns)
    ci_lst = np.zeros(max_expruns)
    
    for run_num in range(max_expruns):
        print("\n Task 5_2: Experiment {} running...\n".format(run_num))
        history, acc, interval, acc_train, acc_test, y_pred = keras_nn(X_train, X_test, y_train, y_test, hidden, outputs, model_type, learn_rate)
        
        fname = 'adam_vs_sgd_2'
        result_present(y_test, y_pred, history, acc, interval, acc_train, acc_test, task, run_num, fname)
    
        acc_lst[run_num] = acc
        ci_lst[run_num] = interval
    
    np.savetxt('data/classification/task5_2_acc.csv', acc_lst, delimiter=',')
    np.savetxt('data/classification/task5_2_CI.csv', ci_lst, delimiter=',')
    
    print("Task5_2 Accuracy score: \n", acc_lst)
    print("\nTask5_2 CI: \n", ci_lst)
    
    acc_mean = np.mean(acc_lst)
    acc_std = np.std(acc_lst)

    ci_mean = np.mean(ci_lst)
    ci_std = np.std(ci_lst)
    
    np.savetxt('data/classification/task5_2_results.txt', (acc_mean, acc_std, ci_mean, ci_std), fmt='%1.5f')

    print("\n Task5_2 Mean Acc: ", acc_mean)
    print("\n Task5_2 std of Acc: ", acc_std)
    print("\n Task5_2 Mean CI: ", ci_mean)
    print("\n Task5_2 std of CI: ", ci_std)


if __name__ == '__main__':
    main()
```
